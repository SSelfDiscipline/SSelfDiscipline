```python
def xywh_to_xywh(size,box):
    """将边界框坐标从VOC格式转换为YOLO格式"""
    dw = 1. / size[0]
    dh = 1. / size[1]
    x = (box[0] + box[2]) / 2
    y = (box[1] + box[3]) / 2
    w = box[2] - box[0]
    h = box[3] - box[1]
    # 保留6位小数
    x = round(x * dw,6)
    w = round(w * dw,6)
    y = round(y * dh,6)
    h = round(h * dh,6)
    return (x, y, w, h)
```
